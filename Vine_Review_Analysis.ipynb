```python
cleaned_vine_df = vine_table_df.loc[(vine_table_df["total_votes"] >= 20)]
cleaned_vine_df
```

```python
cleaned_votes_df = cleaned_vine_df.loc[(cleaned_vine_df["helpful_votes"] / cleaned_vine_df["total_votes"] >= 0.5)]
cleaned_votes_df
```

```python
paid_df = cleaned_votes_df.loc[(cleaned_votes_df["vine"] == "Y")]
paid_df
```

```python
unpaid_df = cleaned_votes_df.loc[(cleaned_votes_df["vine"] == "N")]
unpaid_df
```

```python
fiveStar_paid = paid_df.loc[paid_df["star_rating"] == 5.0]
paid5Star_count = fiveStar_paid.shape[0] 
paid5Star_count
```
